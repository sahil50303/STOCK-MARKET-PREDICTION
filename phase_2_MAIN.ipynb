```python
import pandas as pd

def calculate_pivot_points(df):
    pivot_points = []
    top_pivot_line = []
    bottom_pivot_line = []  # Initialize as a list
    for date, group in df.groupby(df['Timestamp'].dt.date):
        if len(group) >= 125:
            prev_day_data = df[df['Timestamp'].dt.date == date - pd.Timedelta(days=1)].tail(125)
            if len(prev_day_data) == 125:
                pivot_point = (prev_day_data['High'].max() + prev_day_data['Low'].min() + prev_day_data['Close'].iloc[-1]) / 3
                pivot_points.extend([pivot_point] * len(group))
                
                bottom_pivot_line_value = (prev_day_data["High"].max() + prev_day_data["Low"].min()) / 2
                bottom_pivot_line.extend([bottom_pivot_line_value] * len(group))  # Extend the list with the value
            else:
                pivot_points.extend([None] * len(group))
                bottom_pivot_line.extend([None] * len(group))
        else:
            pivot_points.extend([None] * len(group))
            bottom_pivot_line.extend([None] * len(group))

    df['Pivot Point'] = pivot_points
    df['bottom_pivot_line'] = bottom_pivot_line

    return df

# Call the function with your DataFrame 'df'
df = calculate_pivot_points(df)
```

```python
import pandas as pd

def pivot_getting(data):
    res1 = []
    sup1 = []
    res2 = []
    sup2 = []
    top_pivot_line = []
    res3 = []
    sup3 = []
    
    for date, group in data.groupby(data['Timestamp'].dt.date):
        if len(group) >= 125:
            prev_day_data = data[data['Timestamp'].dt.date == date - pd.Timedelta(days=1)].tail(125)
            present_data_data = data[data['Timestamp'].dt.date == date - pd.Timedelta(days=0)].tail(125)
            if len(prev_day_data) == 125:
                res1_val = (2 * present_data_data["Pivot Point"].max()) - prev_day_data['Low'].min()         
                sup1_val = (2 * present_data_data["Pivot Point"].max()) - prev_day_data['High'].max()
                res2_val = present_data_data["Pivot Point"].max() + (prev_day_data['High'].max()-prev_day_data['Low'].min())
                sup2_val = present_data_data["Pivot Point"].max() - (prev_day_data['High'].max()-prev_day_data['Low'].min())
                top_val =  present_data_data["Pivot Point"].max() + (present_data_data["Pivot Point"].max() - present_data_data["bottom_pivot_line"].max())             
                res1.extend([res1_val] * len(group))
                sup1.extend([sup1_val] * len(group))
                res2.extend([res2_val] * len(group))
                sup2.extend([sup2_val] * len(group))
                top_pivot_line.extend([top_val] * len(group))
            else:
                res1.extend([None] * len(group))
                sup1.extend([None] * len(group))
                res2.extend([None] * len(group))
                sup2.extend([None] * len(group))
                top_pivot_line.extend([None] * len(group))
        else:
            res1.extend([None] * len(group))
            sup1.extend([None] * len(group))
            sup2.extend([None] * len(group))
            res2.extend([None] * len(group))
            top_pivot_line.extend([None] * len(group))
            
    data['res1'] = res1
    data['sup1'] = sup1
    data['res2'] = res2
    data['sup2'] = sup2
    data["top_pivot_line"] = top_pivot_line
    return data

```

```python
def identify_price_position(df1):
    upper_bound = []
    lower_bound = []
    for i in df1.columns:
        if df1["Open"][i] > df1["res2"][i]:
            upper_bound.append(None)
            lower_bound.append(df1["res2"][i])
        elif df1["Open"][i] < df1["res2"][i] and df1["Open"][i] > df1["res1"][i]:
            upper_bound.append(df1["res2"][i])
            lower_bound.append(df1["res1"][i])
        elif df1["Open"][i] < df1["res1"][i] and df1["Open"][i] > df1["top_pivot_line"][i]:
            upper_bound.append(df1["res1"][i])
            lower_bound.append(df1["top_pivot_line"][i])
        elif df1["Open"][i] < df1["top_pivot_line"][i] and df1["Open"][i]> df1["Pivot Point"][i]:
            upper_bound = df1["top_pivot_line"][i]
            lower_bound = df1["Pivot Point"][i]
        elif df1["Open"][i] < df1["Pivot Point"][i] and df1["Open"][i] > df1["bottom_pivot_line"][i]:
            upper_bound = df1["Pivot Point"][i]
            lower_bound = df1["bottom_pivot_line"][i]
        elif df1["Open"][i] < df1["bottom_pivot_line"][i] and df1["Open"][i] > df1["sup1"][i]:
            upper_bound.append(df1["bottom_pivot_line"][i])
            lower_bound.append(df1["sup1"][i])
        elif df1["Open"][i] < df1["sup1"][i] and df1["Open"][i] > df1["sup2"][i]:
            upper_bound.append(df1["sup1"][i])
            lower_bound.append(df1["sup2"][i])
        else:
            upper_bound.append(df1["sup2"][i])
            lower_bound.append(None)
    return upper_bound,lower_bound
```

```python
def identify_price_position(df1):
    upper_bound = []
    lower_bound = []
    
    for i in range(len(df1)):  # Iterate over indices instead of columns
        if df1["Open"].iloc[i] > df1["res2"].iloc[i]:
            upper_bound.append(None)
            lower_bound.append(df1["res2"].iloc[i])
        elif df1["Open"].iloc[i] < df1["res2"].iloc[i] and df1["Open"].iloc[i] > df1["res1"].iloc[i]:
            upper_bound.append(df1["res2"].iloc[i])
            lower_bound.append(df1["res1"].iloc[i])
        elif df1["Open"].iloc[i] < df1["res1"].iloc[i] and df1["Open"].iloc[i] > df1["top_pivot_line"].iloc[i]:
            upper_bound.append(df1["res1"].iloc[i])
            lower_bound.append(df1["top_pivot_line"].iloc[i])
        elif df1["Open"].iloc[i] < df1["top_pivot_line"].iloc[i] and df1["Open"].iloc[i] > df1["Pivot Point"].iloc[i]:
            upper_bound.append(df1["top_pivot_line"].iloc[i])
            lower_bound.append(df1["Pivot Point"].iloc[i])
        elif df1["Open"].iloc[i] < df1["Pivot Point"].iloc[i] and df1["Open"].iloc[i] > df1["bottom_pivot_line"].iloc[i]:
            upper_bound.append(df1["Pivot Point"].iloc[i])
            lower_bound.append(df1["bottom_pivot_line"].iloc[i])
        elif df1["Open"].iloc[i] < df1["bottom_pivot_line"].iloc[i] and df1["Open"].iloc[i] > df1["sup1"].iloc[i]:
            upper_bound.append(df1["bottom_pivot_line"].iloc[i])
            lower_bound.append(df1["sup1"].iloc[i])
        elif df1["Open"].iloc[i] < df1["sup1"].iloc[i] and df1["Open"].iloc[i] > df1["sup2"].iloc[i]:
            upper_bound.append(df1["sup1"].iloc[i])
            lower_bound.append(df1["sup2"].iloc[i])
        else:
            upper_bound.append(df1["sup2"].iloc[i])
            lower_bound.append(None)
    df["upper_bound"] = upper_bound
    df["lower_bound"] = lower_bound
    return df.tail(20)
```

```python
import heapq
import talib
rsi_div = []
def bearish_divergence_rsi_macd(data, lockback_period=25):
    signals = []
    bearish_divergence_observed = False  # Flag to track bearish divergence
    bearish_divergence_confirmed = False  # Flag to track confirmed bearish divergence
    
    for i in range(len(data)):
        if i >= lockback_period:
            rsi_column = "rsi14"
            rsi_values = data[rsi_column].iloc[i-lockback_period:i]
            price_highs = data['High'].iloc[i-lockback_period:i]
            two_highest_highs = heapq.nlargest(2, zip(price_highs, range(i-lockback_period, i)))
            highest_highs_indexes = [index for _, index in two_highest_highs]
            rsi_values_highest_highs = rsi_values.loc[highest_highs_indexes]
            rsi_corresponding_To_max_price = rsi_values_highest_highs.iloc[0]
            rsi_corresponding_to_second_max = rsi_values_highest_highs.iloc[1]
            if rsi_corresponding_To_max_price < rsi_corresponding_to_second_max:
                if (65 <= rsi_values_highest_highs.min() <= 85 and 65 <= rsi_values_highest_highs.max() <= 85):
                    if abs(highest_highs_indexes[0] - highest_highs_indexes[1]) >= 4:
                        signals.append(-1)
                        bearish_divergence_observed = True
                    else:
                        signals.append(0)
                else:
                    signals.append(0)
            else:
                signals.append(0)
        else:
            signals.append(0)
    data["signals"] = signals
    return data
                             
```
